```python
#1. Chargement des fichiers
df1 = pd.read_excel("../test/small_base_anomyme.xlsx")
df2 = pd.read_excel("../test/small_datamap.xlsx")

# 2. Créer un dictionnaire de dictionnaires
# Exemple : mapping["RS1_R"] = {nan: 'RS1 Genre', 1.0: 'Un homme', 2.0: 'Une femme', 3.0: 'Non binaire'}
mapping = {
    cat: group.set_index("CODE")["FR:L"].to_dict()
    for cat, group in df2.groupby("NAME")
}

# 3. Remplacer automatiquement chaque colonne de df1
for col in df1.columns: # on ignore les première colonne
    if col in mapping:
        df1[col] = df1[col].apply(
            lambda x: mapping[col].get(x, x) if pd.notna(x) else x
        )

# 4. Sauvegarder le fichier final
df1.to_csv("small_final_3.csv", index=False)

print("✅ Fusion terminée ! Le fichier 'small_final_3.csv' contient les noms des produits.")
```
